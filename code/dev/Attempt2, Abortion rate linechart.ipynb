```python
# Imports and universal whatnots
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
%matplotlib inline

path = '../../data/'
files = glob.glob(os.path.join(path, '*'))
```

```python
# Pandas setup
count_df = pd.read_csv(path+'abortion_count.csv', index_col='Unnamed: 0',  sep=',')
rate_df = pd.read_csv(path+'abortion_rates.csv',  index_col='Unnamed: 0', sep=',')

rate_df.head(16)
```

```python
graph = rate_df.plot( kind='line', legend=False)
plt.ylabel('Rate per 100,000 women age 15-44')
plt.xlabel('Year')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
graph.set_ylim([0,20])
plt.show()
```
